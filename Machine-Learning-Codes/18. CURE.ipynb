```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

Generate synthetic data

```python
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)
```

Fit a hierarchical clustering model (CURE approximation)

```python
hierarchical_model = AgglomerativeClustering(n_clusters=3)
y_hierarchical = hierarchical_model.fit_predict(X)
```

Visualize the clusters from the hierarchical clustering (simulating CURE)

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_hierarchical, s=50, cmap='viridis')
plt.title('Agglomerative Clustering (Hierarchical, k=3)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

Show some representative points (mean points from each cluster)

```python
representative_points = np.array([X[y_hierarchical == i].mean(axis=0) for i in range(3)])
print("Representative points from each cluster:")
print(representative_points)
```

Plot representative points on the cluster plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_hierarchical, s=50, cmap='viridis')
plt.scatter(representative_points[:, 0], representative_points[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Mark representative points
plt.title('Agglomerative Clustering with Representative Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

Visualize the dendrogram (Hierarchical Clustering structure)

```python
plt.figure(figsize=(10, 6))
Z = linkage(X, 'ward')
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering (Simulating CURE)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

Calculate essential metrics for evaluation

```python
silhouette_avg_agg = silhouette_score(X, y_hierarchical)  # Mean silhouette score

# Print the metrics
print(f"Agglomerative Clustering Silhouette Score: {silhouette_avg_agg:.2f}")
```
